### Linear regression

```python
from sklearn.linear_model import LinearRegression
X_train,y_train =train_data.drop(['MedHouseVal'],axis=1),train_data['MedHouseVal']
reg=LinearRegression()
```

```python
reg.fit(X_train,y_train)
```

```python
reg.score(X_test,y_test)
```

```python
from sklearn.preprocessing import  StandardScaler
scaler=StandardScaler()
X_train_s=scaler.fit_transform(X_train)
reg.fit(X_train_s,y_train)
```

```python
X_test_s=scaler.fit_transform(X_test)
reg.score(X_test_s,y_test)
```

### Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
forest=RandomForestRegressor()
forest.fit(X_train,y_train)
```

```python
forest.score(X_test,y_test)
```

### GridsearchCV

```python
from sklearn.model_selection import GridSearchCV
forest=RandomForestRegressor()
param_grid= {
          "n_estimators":[3,10,20],
          "max_features":[2,4,6,8]
}

grid_search = GridSearchCV(forest,param_grid,cv=5,scoring="neg_mean_squared_error",return_train_score=True)
grid_search.fit(X_train,y_train)
```

```python
best_forest=grid_search.best_estimator_
```

```python
best_forest.score(X_test,y_test)
```

```python
from sklearn.metrics import mean_squared_error
y_pred=best_forest.predict(X_test)
mse=mean_squared_error(y_test,y_pred)
print(f'MSE is {mse}')
```

```python
results=pd.DataFrame({'Actual Values': y_test,'Predicted values' : y_pred})
print(results)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, color='blue')
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Actual vs Predicted House Prices')

# Draw the ideal line
plt.plot([0, 6], [0, 6], color='red', linestyle='--')

# Set limits to minimize empty space
plt.xlim(y_test.min() - 1, y_test.max() + 1)
plt.ylim(y_test.min() - 1, y_test.max() + 1)

plt.grid()
plt.show()
```

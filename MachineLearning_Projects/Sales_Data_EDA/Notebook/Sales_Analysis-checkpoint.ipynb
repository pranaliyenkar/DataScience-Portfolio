### Q2 Which city has highest number of sales?

#### Add a city Column

```python
all_data.head()
```

```python
import warnings
warnings.filterwarnings('ignore')
def get_city(address):
    return address.split(',')[1]

def get_state(address):
     return address.split(',')[2].split( )[0]

all_data['City']=all_data['Purchase Address'].apply(lambda x : f"{get_city(x)}({get_state(x)})")
#all_data['State']=all_data['Purchase Address'].apply(lambda x : get_state(x).strip())
all_data.head(5)
```

```python
results1=all_data.groupby('City').sum()[['Sales']]
results1
```

```python
cities=all_data['City'].unique()
cities =[city for city ,df in all_data.groupby('City')]

plt.bar(cities,results1['Sales'])
plt.xticks(cities,rotation='vertical',size=8)
plt.xlabel('City')
plt.ylabel('Sales')
plt.show()
```

### Q3: What time we should disply thadvertisement to maximize likelihood of customers buying product?

### Q 4: What products most often sold together

```python
all_data.head()
```

```python
df=all_data[all_data['Order ID'].duplicated(keep=False)]
df.head(10)
```

```python
df['Grouped']=df.groupby('Order ID')['Product'].transform(lambda x : ','.join(x))
df.head()
```

```python
df=df[['Order ID','Grouped']].drop_duplicates()
df.head()
```

```python
r1=df.groupby('Grouped').count().sort_values(by='Order ID',ascending=False).head(10)
r1=r1[['Order ID']]
r1 = r1.reset_index()
r1.index = range(1, len(r1) + 1)
r1.head(10)
```

```python
print(r1)
```

```python
plt.figure(figsize=(6,3))
plt.bar(r1['Grouped'],r1['Order ID'])

plt.xticks(r1['Grouped'],rotation='vertical',size=8)
plt.xlabel('Items Sold Together')
plt.ylabel('Num of Orders')
plt.show()
```
